```python
df_check = pd.read_csv("new_dataframe.csv")
df_check.head()

```

```python
from scipy.stats import ttest_ind
```

```python
# Creating a new column for Overall Mental Health Score
df_check['Overall Mental Health Score'] = df_check[['Anxiety', 'Depression', 'Insomnia', 'OCD']].mean(axis=1)

# Cleaning and filtering data for analysis
df_test = df_check[['Age', 'Overall Mental Health Score', 'While working']].dropna()

# Splitting the data into two groups: Those who listen to music while working and those who don't
group_yes = df_test[df_test['While working'] == 'Yes']['Overall Mental Health Score']
group_no = df_test[df_test['While working'] == 'No']['Overall Mental Health Score']

# Performing an independent T-test to compare means
t_stat, p_value = ttest_ind(group_yes, group_no, equal_var=False)  # Welchâ€™s T-test to account for unequal variances

# Display results
t_stat, p_value

```

### Statistically Significant (p-value < 0.05)
Since the p-value (0.0477) is below 0.05, we reject the null hypothesis, indicating **a significant difference in overall mental health scores between individuals who listen to music while working and those who donâ€™t**.

### Key Takeaways
- This suggests that listening to music while working does have some measurable impact on mental health.

```python
# Recalculate mean and standard deviation values for each group
mean_values = [group_yes.mean(), group_no.mean()]
std_values = [group_yes.std(), group_no.std()]

# Define labels again
labels = ["Listening to Music", "Not Listening to Music"]

# Set up the figure
plt.figure(figsize=(8, 5))

# Create a bar chart
plt.bar(labels, mean_values, yerr=std_values, capsize=5, color=["#1DB954", "#FF5252"])

# Annotate the t-test result and p-value on the chart
t_stat_text = f"t-statistic = {round(t_stat, 2)}\np-value = {round(p_value, 3)}"
plt.text(0.5, max(mean_values) - 0.5, t_stat_text, ha="center", fontsize=12, bbox=dict(facecolor="white", alpha=0.5))

# Add labels and title
plt.xlabel("Group", fontsize=12)
plt.ylabel("Average Mental Health Score", fontsize=12)
plt.title("T-Test Results: Music & Mental Health", fontsize=14)
plt.ylim(0, max(mean_values) + 2)  # Adjust y-axis

# Display the plot
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
